```
with open('/content/drive/My Drive/task_retai/images_variant_train.txt', 'r') as document:
    variant_dict = {}
    for line in document:
        line = line.split()
        if not line:  # empty line?
            continue
        variant_dict[line[0]] = line[1:]
```

```
for i in variant_dict:
    if len(variant_dict[i])==1:
        variant_dict[i]=str(variant_dict[i][0])
        
    if len(variant_dict[i])==2:
        variant_dict[i]=str(variant_dict[i][0]+" "+variant_dict[i][1])
        
    if len(variant_dict[i])==3:
        variant_dict[i]=str(variant_dict[i][0]+" "+variant_dict[i][1]+" "+variant_dict[i][2])
        
```

```
for i in family_dict:
    if len(family_dict[i])==1:
        family_dict[i]=str(family_dict[i][0])
    if len(family_dict[i])==2:
        family_dict[i]=str(family_dict[i][0]+" "+family_dict[i][1])
        
    if len(family_dict[i])==3:
        family_dict[i]=str(family_dict[i][0]+" "+family_dict[i][1]+" "+family_dict[i][2])
```

```
dict_final={}
for key,value in variant_dict.items():
    if value not in dict_final.keys():
        dict_final[value] = family_dict[key]
```

```
for i, cls_index in enumerate(class_indices):
  image_paths = np.array(glob(f'{train_folders[cls_index]}/*.jpg'))
  class_name = class_names[i]
  print(f'{class_name}: {len(image_paths)}')
  np.random.shuffle(image_paths)
  train,val,test = np.split(image_paths,indices_or_sections=[int(.8*len(image_paths)), int(.9*len(image_paths))])

 
  for ip in train :

    shutil.copy(ip, f'data/train/{class_name}/')
  for ip in val:
    shutil.copy(ip, f'data/val/{class_name}/')
  for ip in test:
    shutil.copy(ip, f'data/test/{class_name}/')
```
